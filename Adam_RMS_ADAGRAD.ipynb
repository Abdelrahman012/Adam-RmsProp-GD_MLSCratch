```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as m
```

```python
# Linear Regression Model
import numpy as np
from numpy.linalg import norm
# h(x)=theta_0+theta_1*x
x_points=np.linspace(0,20,50).reshape(50,1)
y_points=-2*x_points+1
#hypothes to equation with theta_o value=10 and  theta_1=10

def calc(wt_0,wt_1):
    y_pred=wt_0+wt_1*x_points

    loss_function=1/ (2*y_pred.shape[0])*((y_pred-y_points)**2).sum()    
    return (loss_function,y_pred)
    
def Adagrad(x_points,y_points,alpha,epison,max_iter):
    wt_0=1
    wt_1=1
    loss_function,y_pred=calc(wt_0,wt_1)

    iter=0
    iter_lst=[]
    loss_function_lst=[]
    wt0_lst=[]
    wt1_lst=[]
    vt0=0
    vt1=0
    epison=epison
    while iter<max_iter:
        alpha=alpha
        wt0_lst.append(wt_0) #make a list of theta zer0 to make our plot with loss function
        wt1_lst.append(wt_1) #make a list of thta one to make our plot
        ###################
        gradient_wt0=(1/y_pred.shape[0])*((y_pred-y_points).sum())
        gradient_wt1=(1/y_pred.shape[0])*((y_pred-y_points)*x_points).sum()
        #########################3
        vt0=vt0+(gradient_wt0)**2
        vt1=vt1+(gradient_wt1)**2
        #############################
        wt_0=wt_0-(alpha/(m.sqrt(vt0)+epison))*gradient_wt0
        wt_1=wt_1-(alpha/(m.sqrt(vt1)+epison))*gradient_wt1
        loss_function,y_pred=calc(wt_0,wt_1)
        ######################3
        iter_lst.append(iter)
        loss_function_lst.append(loss_function)
        #############################3A
        wt0_lst.append(wt_0)
        wt1_lst.append(wt_1)
        ############################ 
        
        iter+=1
        if loss_function<.001:
            break
    return wt_0,wt_1,loss_function,wt0_lst,wt1_lst,loss_function_lst,iter
wt_0,wt_1,loss_function,lst0,lst1,lst3,iteration=Adagrad(x_points,y_points,.1,1e-9,2000)
print(f"theta_0 is {theta_0},theta_1 is {theta_1},loss_function is {loss_function}")    
print(iteration)
y_pred1=wt_0+wt_1*x_points
print(y_pred1)
```

# RMSProp

```python
# Linear Regression Model
import numpy as np
from numpy.linalg import norm
# h(x)=theta_0+theta_1*x
x_points=np.linspace(0,20,50).reshape(50,1)
y_points=-2*x_points+1
#hypothes to equation with theta_o value=10 and  theta_1=10

def calc(wt_0,wt_1):
    y_pred=wt_0+wt_1*x_points

    loss_function=1/ (2*y_pred.shape[0])*((y_pred-y_points)**2).sum()    
    return (loss_function,y_pred)
    
def RMSProp(x_points,y_points,alpha,epison,bita,max_iter):
    wt_0=1
    wt_1=1
    loss_function,y_pred=calc(wt_0,wt_1)

    iter=0
    iter_lst=[]
    loss_function_lst=[]
    wt0_lst=[]
    wt1_lst=[]
    vt0=0
    vt1=0
    epison=epison
    bita=bita
    while iter<max_iter:
        alpha=alpha
        wt0_lst.append(wt_0) #make a list of theta zer0 to make our plot with loss function
        wt1_lst.append(wt_1) #make a list of thta one to make our plot
        ###################
        gradient_wt0=(1/y_pred.shape[0])*((y_pred-y_points).sum())
        gradient_wt1=(1/y_pred.shape[0])*((y_pred-y_points)*x_points).sum()
        #########################3
        vt0=bita*vt0+(1-bita)*(gradient_wt0)**2
        vt1=bita*vt1+(1-bita)*(gradient_wt1)**2
        #############################
        wt_0=wt_0-(alpha/(m.sqrt(vt0)+epison))*gradient_wt0
        wt_1=wt_1-(alpha/(m.sqrt(vt1)+epison))*gradient_wt1
        loss_function,y_pred=calc(wt_0,wt_1)
        ######################3
        iter_lst.append(iter)
        loss_function_lst.append(loss_function)
        #############################3A
        wt0_lst.append(wt_0)
        wt1_lst.append(wt_1)
        ############################ 
        
        iter+=1
        if loss_function<.001:
            break
    return wt_0,wt_1,loss_function,wt0_lst,wt1_lst,loss_function_lst,iter
wt_0,wt_1,loss_function,lst0,lst1,lst3,iteration=RMSProp(x_points,y_points,.1,1e-9,.98,2000)
print(f"theta_0 is {theta_0},theta_1 is {theta_1},loss_function is {loss_function}")    
print(iteration)
y_pred1=wt_0+wt_1*x_points
print(y_pred1)
```

```python
from sklearn.metrics import r2_score
R2=r2_score(y_true=y_points,y_pred=y_pred1)
print(R2)
```

# Adam

```python
# Linear Regression Model
import numpy as np
from numpy.linalg import norm
# h(x)=theta_0+theta_1*x
x_points=np.linspace(0,20,50).reshape(50,1)
y_points=-2*x_points+1
#hypothes to equation with theta_o value=10 and  theta_1=10

def calc(wt_0,wt_1):
    y_pred=wt_0+wt_1*x_points

    loss_function=1/ (2*y_pred.shape[0])*((y_pred-y_points)**2).sum()    
    return (loss_function,y_pred)
    
def ADAM(x_points,y_points,alpha,epison,bita1,bita2,max_iter):
    wt_0=1
    wt_1=1
    loss_function,y_pred=calc(wt_0,wt_1)

    iter=0
    iter_lst=[]
    loss_function_lst=[]
    wt0_lst=[]
    wt1_lst=[]
    
    vt0=0
    vt1=0
    
    mt0=0
    mt1=0
    
    epison=epison
    bita1=bita1
    bita2=bita2
    iter=1
    
    
    while iter<max_iter:
        alpha=alpha
        wt0_lst.append(wt_0) #make a list of theta zer0 to make our plot with loss function
        wt1_lst.append(wt_1) #make a list of thta one to make our plot
        #############################
        gradient_wt0=(1/y_pred.shape[0])*((y_pred-y_points).sum())
        gradient_wt1=(1/y_pred.shape[0])*((y_pred-y_points)*x_points).sum()
        #############################
        mt0=bita1*mt0+(1-bita1)*gradient_wt0
        mt1=bita1*mt1+(1-bita1)*gradient_wt1
        #############################
        vt0=bita2*vt0+(1-bita2)*(gradient_wt0)**2
        vt1=bita2*vt1+(1-bita2)*(gradient_wt1)**2
        #############################
       
        ########################
        wt_0=wt_0-(alpha/(m.sqrt(vt0)+epison))*mt0
        wt_1=wt_1-(alpha/(m.sqrt(vt1)+epison))*mt1
        loss_function,y_pred=calc(wt_0,wt_1)
        ######################3
        iter_lst.append(iter)
        loss_function_lst.append(loss_function)
        #############################3A
        wt0_lst.append(wt_0)
        wt1_lst.append(wt_1)
        ############################ 
        
        iter+=1
        mt0=mt0/(1-bita1**iter)
        mt1=mt1/(1-bita1**iter)
        
        vt0=vt0/(1-bita2**iter)
        vt1=vt1/(1-bita2**iter)
        if loss_function<.001:
            break
    return wt_0,wt_1,loss_function,wt0_lst,wt1_lst,loss_function_lst,iter
wt_0,wt_1,loss_function,lst0,lst1,lst3,iteration=ADAM(x_points,y_points,.01,1e-8,.9,.9,10000)
print(f"theta_0 is {theta_0},theta_1 is {theta_1},loss_function is {loss_function}")    
print(iteration)
y_pred1=wt_0+wt_1*x_points
print(y_pred1)
 
```
